## DATA EXTRACTION

## Data Extraction - Data-> Aggregated -> Transaction -> Country -> India -> State

```python
# Data Extraction  --> data->aggregated->transaction->country->India->State

#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

# This is to direct the path to get the data as states
path1 = "C:\\Users\\HP\\Desktop\\Phonepe_Pulse\\pulse\\data\\aggregated\\transaction\\country\\india\\state\\"
Agg_state_list = os.listdir(path1)
# Agg_state_list--> to get the list of states in India

# <------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------->

# This is to extract the data and create a dataframe
clm1 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

for i in Agg_state_list:
    p_i = path1 + i + "\\"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i + j + "\\"
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D1 = json.load(Data)
                # Iterate through 'transactionData' in the JSON
                for z in D1['data']['transactionData']:
                    Name = z['name']  # Extract the transaction name/category (e.g., Recharge & bill payments)
                    
                    # Extract the details of each payment instrument for a transaction in the transactionData array of the JSON structure
                    # Extract the details from 'paymentInstruments'
                    for instrument in z['paymentInstruments']:
                        count = instrument['count']  # Extract the count of transactions
                        amount = instrument['amount']  # Extract the total amount

                        # Append the extracted data to the respective lists in the dictionary
                        clm1['Transaction_type'].append(Name)
                        clm1['Transaction_count'].append(count)
                        clm1['Transaction_amount'].append(amount)
                        clm1['State'].append(i)
                        clm1['Year'].append(j)
                        clm1['Quarter'].append(int(k.strip('.json')))

# Successfully created a dataframe
Aggregated_Transaction = pd.DataFrame(clm1)

# Print the first few rows of the DataFrame to see the result
Aggregated_Transaction.head()
```

## Data Extraction - data-> Aggregated -> User -> Country -> India -> State

```python
# Path to aggregated user data for states
path2 = "C:\\Users\\HP\\Desktop\\Phonepe_Pulse\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Agg_state_list = os.listdir(path2)

# Initialize a dictionary to hold the extracted data
clm2 = {'State': [], 'Year': [], 'Quarter': [], 'Brand': [], 'user_count_by_brand': [], 'Percentage': [], 'Registered_Users': [], 'App_Opens': []}

for i in Agg_state_list:
    p_i = path2 + i + "\\"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i + j + "\\"
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D2 = json.load(Data)

                # Get aggregated data for registered users and app opens
                if D2['data'].get('aggregated'):
                    registered_users = D2['data']['aggregated']['registeredUsers']
                    app_opens = D2['data']['aggregated']['appOpens']

                # Check if 'usersByDevice' exists and is not None
                if D2['data'].get('usersByDevice'):
                    for z in D2['data']['usersByDevice']:
                        brand = z['brand']  # Brand name of the device
                        count = z['count']  # Number of registered users by this brand
                        percentage = z['percentage']  # Percentage share of current device type

                        # Append the data to the dictionary
                        clm2['Brand'].append(brand)
                        clm2['user_count_by_brand'].append(count)
                        clm2['Percentage'].append(percentage)
                        clm2['State'].append(i)
                        clm2['Year'].append(j)
                        clm2['Quarter'].append(int(k.strip('.json')))  # Convert quarter to integer
                        clm2['Registered_Users'].append(registered_users)
                        clm2['App_Opens'].append(app_opens)

# Create a Pandas DataFrame from the dictionary
Aggregated_Users = pd.DataFrame(clm2)

# Print the first few rows of the DataFrame to see the result
Aggregated_Users.head(30)

```

## Data Extraction - data -> Aggregated -> Insurance -> Country - > India -> State

```python
# Path to aggregated insurance data for states
path3 = "C:\\Users\\HP\\Desktop\\Phonepe_Pulse\\pulse\\data\\aggregated\\insurance\\country\\india\\state\\"
Agg_state_list = os.listdir(path3)

# Initialize a dictionary to hold the extracted data
clm3 = {'State': [], 'Year': [], 'Quarter': [], 'Insurance_type': [], 'Insurance_count': [], 'Insurance_amount': []}

# Loop through each state
for i in Agg_state_list:
    p_i = path3 + i + "\\"
    Agg_yr = os.listdir(p_i)

    # Loop through each year
    for j in Agg_yr:
        p_j = p_i + j + "\\"
        Agg_yr_list = os.listdir(p_j)

        # Loop through each quarter
        for k in Agg_yr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D3 = json.load(Data)

                # Check if 'transactionData' exists and is not empty
                if D3['data'].get('transactionData'):
                    # Since there is only one entry for 'transactionData', directly access it
                    transaction = D3['data']['transactionData'][0]  # Only one item

                    # Extract the transaction type ("Insurance")
                    Name = transaction['name']  # In this case, it's always "Insurance"

                    # Extract the first (and only) payment instrument
                    instrument = transaction['paymentInstruments'][0]  # Single entry
                    count = instrument['count']  # Number of insurance transactions
                    amount = instrument['amount']  # Total value of insurance transactions

                    # Append data to the dictionary
                    clm3['Insurance_type'].append(Name)
                    clm3['Insurance_count'].append(count)
                    clm3['Insurance_amount'].append(amount)
                    clm3['State'].append(i)
                    clm3['Year'].append(j)
                    clm3['Quarter'].append(int(k.strip('.json')))  # Ensure the quarter is an integer

# Successfully create a dataframe
Aggregated_insurance = pd.DataFrame(clm3)

# Print the first few rows of the DataFrame
Aggregated_insurance.head(30)
```

```python

# This is to direct the path to get the data as states
path4 = "C:\\Users\\HP\\Desktop\\Phonepe_Pulse\\pulse\\data\\map\\transaction\\hover\\country\\india\\state\\"
Map_state_list = os.listdir(path4)

# Initialize a dictionary to hold the extracted data
clm4 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

# Loop through each state
for i in Map_state_list:
    p_i = path4 + i + "\\"
    Map_yr = os.listdir(p_i)
    
    # Loop through each year
    for j in Map_yr:
        p_j = p_i + j + "\\"
        Map_yr_list = os.listdir(p_j)
        
        # Loop through each quarter
        for k in Map_yr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D4 = json.load(Data)

                # Loop through hoverDataList to extract district/state-level data
                for z in D4['data']['hoverDataList']:
                    # Extract the state/district name
                    Name = z['name']

                    # Loop through 'metric' to extract count and amount for each type
                    for metric in z['metric']:
                        count = metric['count']  # Total number of transactions
                        amount = metric['amount']  # Total transaction amount

                        # Append the extracted data to the dictionary
                        clm4['Transaction_type'].append(metric['type'])  # Extract 'type' (e.g., TOTAL)
                        clm4['Transaction_count'].append(count)
                        clm4['Transaction_amount'].append(amount)
                        clm4['State'].append(Name)  # State/District Name
                        clm4['Year'].append(j)
                        clm4['Quarter'].append(int(k.strip('.json')))  # Remove the '.json' extension from quarter

# Successfully create a dataframe
Map_Transaction = pd.DataFrame(clm4)

# Print the first 20 rows of the DataFrame
Map_Transaction.head(30)
```
